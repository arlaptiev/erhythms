```python
def bjorklund(steps, pulses):
    steps = int(steps)
    pulses = int(pulses)
    if pulses > steps:
        raise ValueError    
    pattern = []    
    counts = []
    remainders = []
    divisor = steps - pulses
    remainders.append(pulses)
    level = 0
    while True:
        counts.append(divisor // remainders[level])
        remainders.append(divisor % remainders[level])
        divisor = remainders[level]
        level = level + 1
        if remainders[level] <= 1:
            break
    counts.append(divisor)
    
    def build(level):
        if level == -1:
            pattern.append(0)
        elif level == -2:
            pattern.append(1)         
        else:
            for i in range(0, counts[level]):
                build(level - 1)
            if remainders[level] != 0:
                build(level - 2)
    
    build(level)
    i = pattern.index(1)
    pattern = pattern[i:] + pattern[0:i]
    return pattern
```

```python
table = [[[0] for i in range(65)] for i in range(65)]
```

```python
for i in range(1, 65):
  for j in range(1, i+1):
    table[i][j] = bjorklund(i, j)
```

```python
def binary_list(list_of_1s_and_0s):
  binary_number = ""
  for bit in list_of_1s_and_0s:
    binary_number += str(bit)
  return binary_number

for i in range(len(table)):
  for j in range(len(table[i])):
    table[i][j] = binary_list(table[i][j])

table
```
